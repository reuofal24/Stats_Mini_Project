## 2. Descriptive Statistics:âœ…
* Compute basic statistics (mean, median, mode, standard deviation) for relevant columns (price).

```python
# Calculate the mean order price
mean_order_price = df.groupby('order_id')['price'].sum().mean()
print("Mean order price:", mean_order_price)
```

```python
# Calculate the median order price
median_order_price = df.groupby('order_id')['price'].sum().median()
print("Median order price:", median_order_price)
```

```python
# Calculate the standard deviation order price
std_order_price = df.groupby('order_id')['price'].sum().std()
print("Standard deviation of order prices:", std_order_price)
```

```python
# Calculate the mode order price
mode_order_price = df.groupby('order_id')['price'].sum().mode().iloc[0]
print("Mode order price:", mode_order_price)
```

## 3. Hypothesis Testing (A/B Testing):âœ…
* Hypothesis:
The average order price (amount spent on orders) is the same for orders with different order states (e.g., delivered vs. canceled).

* Null Hypothesis (H0):
There is no difference in the average order price between different order states.

* Alternative Hypothesis (H1):
There is a difference in the average order price between different order states.

```python
from scipy.stats import f_oneway

# Extract order prices for each order state
order_prices_delivered = merged_df[merged_df['order_status'] == 'delivered']['price']
order_prices_canceled = merged_df[merged_df['order_status'] == 'canceled']['price']
order_prices_processing = merged_df[merged_df['order_status'] == 'processing']['price']

# Perform ANOVA test
f_statistic, p_value = f_oneway(order_prices_delivered, order_prices_canceled, order_prices_processing)

# Print results
print("F-statistic:", f_statistic)
print("P-value:", p_value)

# Interpret results
alpha = 0.05
if p_value < alpha:
    print("Reject null hypothesis: There is a statistically significant difference in the average order prices between different order states.")
else:
    print("Fail to reject null hypothesis: There is no statistically significant difference in the average order prices between different order states.")
```
